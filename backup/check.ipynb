```python
import pandas as pd

counter = 0
df_total = pd.DataFrame()
for date in pd.date_range(start='2020/07/01', end='2020/10/01'):
    #print(date.strftime('%Y_%m_%d'))
    df = pd.read_pickle(f"/opt/workspace/persistent-storage/crab-snaps/{date.strftime('%Y_%m_%d')}.pkl")
    df['date'] = date
    df['DATA_TIER'] = df['DESIRED_CMSDataset'].str.rsplit("/").str[-1]
    df_total = pd.concat([df_total, df]).sort_values(by=['date'])
        
    df_total.drop_duplicates(subset=['DESIRED_CMSDataset', "Workflow"], keep='first', inplace=True)
```
